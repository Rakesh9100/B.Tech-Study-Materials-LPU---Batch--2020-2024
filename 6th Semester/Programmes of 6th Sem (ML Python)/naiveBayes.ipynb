```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
data=df.drop(columns=['id','Unnamed: 32'],axis=1)
#print(data.diagnosis.unique())
#print(data.info())
#print(data.isna().sum())
diag={"M":1,"B":0}
data.diagnosis=data.diagnosis.map(diag)
```

```python
X=data.drop(columns=['diagnosis'],axis=1)
Y=data.diagnosis
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3)
```

```python
M=data[data.diagnosis==1]
B=data[data.diagnosis==0]
plt.title("Malignant vs benign")
plt.xlabel("radius mean")
plt.ylabel("texture mean")
plt.scatter(M.radius_mean, M.texture_mean,color="red",label="Malignant",alpha=0.5)
plt.scatter(B.radius_mean, B.texture_mean,color="yellow",label="Malignant",alpha=0.5)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
print("accuracy score for testing ",accuracy_score(ytest,model.predict(xtest)))
print("accuracy score for training ",accuracy_score(ytrain,model.predict(xtrain)))
```
