```python
import numpy as np
import pandas as pd
df=pd.read_csv("salary_data.csv")
#print(df)
import matplotlib.pyplot as plt
plt.title("Experience and salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.scatter(df['YearsExperience'],df['Salary'])
plt.show()
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df['YearsExperience'],df['Salary'])
print(xtrain.shape,xtest.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain.values.reshape(-1,1),ytrain)
```

```python
pred=lr.predict(xtest.values.reshape(-1,1))
pred_train=lr.predict(xtrain.values.reshape(-1,1))
plt.title("Experience and salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.scatter(xtrain,ytrain)
plt.plot(xtrain,pred_train)
plt.show()
print("Mean absolute Error ",np.sum(np.abs(ytrain-pred_train))/ytrain.shape[0])
```

```python
plt.title("Experience and salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.scatter(xtest,ytest)
plt.plot(xtest,pred)
plt.show()
```

```python
df=pd.read_csv("position_salaries.csv")
#print(df)
'''plt.scatter(df['Level'],df['Salary'])
plt.show()'''
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(df['Level'].values.reshape(-1,1))
lr=LinearRegression()
lr.fit(X_poly,df['Salary'])
pred=lr.predict(X_poly)
plt.scatter(df['Level'],df['Salary'])
plt.plot(df['Level'],pred)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import datasets
data=datasets.load_diabetes()
x=pd.DataFrame(data.data,columns=data.feature_names)
y=data.target
#print(x.head())
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3)
dtr=DecisionTreeRegressor()
dtr.fit(xtrain,ytrain)
pred=dtr.predict(xtest)
plt.plot(ytest,color='red')
plt.plot(pred,color='green')
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(ytest,pred))
```
